## 파일 읽고 쓰기
### 1) 파일 생성하기 : open()

```python
f = open('newfile.txt', 'w')
f.close()
```

```python
f = open('newfile.txt', 'w')

for i in range(1,11):
    data = '%d번째 줄입니다.\n'%i
    f.write(data)
f.close()
```

## 3) 파일을 읽는 여러가지 방법

- readline() 함수 이용하기

```python
f = open('newfile.txt','r')
line = f.readline() # 한 줄씩 읽어오는 함수
print(line)
f.close()
```

- readline()함수를 이용해서 여러 줄 읽기

```python
f = open('newfile.txt','r')

while True:
    line = f.readline()
    if not line:
        break
    print(line)

f.close()
```

- 사용자의 입력을 받아 여러 줄 출력

```python
while True:
    data = input()
    if not data:
        break
    print(data)
```

- readlines() 함수 사용하기

```python
f = open('newfile.txt','r')
lines = f.readlines()
lines
```

```python
for line in lines:
    print(line)
f.close()
```

- 줄바꿈(\n)문자 제거 : strip()

```python
f = open('newfile.txt','r')
lines = f.readlines()

for line in lines:
    line = line.strip() #줄바꿈 문자 제거
    print(line)

f.close()
```

- read() 함수 사용하기 : 파일 내용 전체를 문자열로 반환

```python
f = open('newfile.txt','r')
data = f.read()
data
```

```python
print(data)
```

- 파일 객체를 for문으로 읽어오기

```python
with open('foo.txt','w') as f:
    f.write('Life is too short, you need python')
```

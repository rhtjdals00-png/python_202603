## 튜플형 가변 매개변수(입력값을 여러 개 받는 함수)

```python
def add_many(*args):
    result=0
    for i in args:
        result = result + i
    return result

calc = add_many(1,5,9)
print(calc)
```

```python
def add_mul(choice, *args):
    if choice == 'add':
        result = 0
        for i in args:
            result += i
    elif choice == 'mul':
        result = 1
        for i in args:
            result *= i
    return result
```

```python
add = add_mul('add', 1,2,3,4,5)
print(add)
```

```python
mul = add_mul('mul', 1,2,3,4,5)
print(mul)
```

## 함수 안에서 함수 밖의 변수를 변경하는 방법

### 1) return 사용하기

```python
a = 1 #전역 변수
def vartest(a):
    a = a + 1
    return a

a = vartest(a) #전역 변수에 함수의 리턴값을 덮어씀
print(a)
```

### 2) global 명령어 사용하기

```python
a = 1
def vartest():
    global a #전역변수로 변경
    a += 1

vartest()
print(a)
```

## 함수의 활용

### 1) 재귀 함수 : 자기 자신(함수)을 호출하는 함수

```python
def factorial(n):
    output = 1
    for i in range(1, n+1):
        output *= i
    return output
```

```python
print('1! :', factorial(1))
print('2! :', factorial(2))
print('3! :', factorial(3))
print('4! :', factorial(4))
print('5! :', factorial(5))
```

```python
# 재귀함수로 팩토리얼 구하기
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
print('5! = ', factorial(5))
```

## 2) 리스트를 평탄화하는 재귀 함수 만들기

```python
# for문으로 평탄화(flatten)

def flatten(data):
    output = []
    for item in data:
        if type(item) == list:
            output += item
        else:
            output.append(item)
    return output
```

```python
# 재귀함수로 리스트 평탄화
def flatten(data):
    output = []
    for item in data:
        if type(item) == list:
            output += flatten(item)
        else:
            output.append(item)
    return output
```

```python
print('원본 data : ', ex)
```

```python
print('변환 data: ' , flatten(ex))
```

## 1. 계산기 프로그램을 만들며 클래스 알아보기

```python
result = 0 #초기값

def add(num):
    global result
    result += num
    return result

print(add(3))
print(add(4))
```

```python
result
```

```python
# 2대의 계산기
result1 = 0
result2 = 0

def add1(num):
    global result1
    result1 += num
    return result1

def add2(num):
    global result2
    result2 += num
    return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))
```

```python
# 여러대 계산기를 클래스로 해결하기
class Calculator: # 클래스 = 설계도
    def __init__(self):
        self.result = 0 #초기값 설정

    def add(self, num):
        self.result += num
        return self.result
```

```python
cal1 = Calculator()
cal2 = Calculator()
```

```python
print(cal1.add(3))
```

```python
print(cal1.add(4))
```

```python
print(cal2.add(3))
```

```python
print(cal2.add(7))
```

```python
# 4) 곱하기, 빼기, 나누기 기능 만들기
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result

    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal()
a.setdata(4,2)
a.add()
```

```python
a.sub()
```

```python
a.mul()
```

```python
a.div()
```

```python
b = FourCal()
b.setdata(3,8)
```

```python
b.add()
```

```python
b.mul()
```

```python
b.div()
```

## 4. 생성자(Constructor)

- 생성자는 객체가 생성될 때 자동으로 실행되는 메서드
- 객체에 초기값을 설정하는 역할

```python
class FourCal:
    # 생성자 함수 : 객체 생성시 자동 호출
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result

    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal()
```

```python
a = FourCal(4,2)
```

```python
a.add()
```

```python
a.sub()
```

```python
a.mul()
```

```python
a.div()
```

```python
a.first
```

```python
a.second
```

## 6. 메서드 오버라이딩(메서드 재정의)

```python
# 0으로 나눌 때 오류가 아닌 0을 리턴하도록 처리하려면?

class SafeFourCal(FourCal):
    def div(self):  #메서드 오버라이딩 : 상속받는 클래스의 메서드 이름을 똑같이 사용하여 함수를 재정의
        if self.second == 0:
            return 0
        else:
            return self.first / self.second
```
